# Train-test splitting
### Training data should represent all set of vals not only some specific ones. Eg if only CHAS=1 is fed in all training data, the model would learn that CHAS=0 is not possible. So, Stratified shuffle split is needed in case of important parameters.
### np.random.seed(42) is used to limit data set to one set only else more compilation would overtrain the model by exposing it to all datapoints.

```python
#def split(data, test_ratio):
#    np.random.seed(42)
#    shuff = np.random.permutation(len(data))
#    test_size = int(len(data)*test_ratio)
#    test_indices = shuff[:test_size]
#    train_indices = shuff[test_size:]
#    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
#train_set , test_set = split(house, 0.2) --> commented as now we need stratified shuffling
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
splitted = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in splitted.split(house, house['CHAS']):
    
    #splitted.split --> this split is func of stratified split class and not func writted above.
    
    strat_train_set = house.loc[train_index]
    strat_test_set = house.loc[test_index]
```

```python
house = strat_train_set.drop("MEDV", axis=1)    #removing actual prices from data set
house_labels = strat_train_set["MEDV"].copy()   #storing actual prices in sepaerate variables
```

# Missing attributes case
## 3 optns available:

> 1. Get rid of missing datapts.
> 2. Get rid of whole attribute (only if that feature is not important).
> 3. Set val to some val (0 or mean or median) as per convinence. Then fit that val in test set too.

```python
#Opt 1
house.shape
#Remove those vals
a = house.dropna(subset=["RM"])
a.describe()
#Original data frame is unch unless we use inplace=True
```

```python
#Opt 2
house.drop("RM",axis=1).describe()
```

```python
#Opt 3
medianRM = house["RM"].median()
print(medianRM)
house["RM"].fillna(medianRM)
```

```python
#Before filling missing vals
house.describe()
```

```python
#Set val in test set
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy = "median")
imp.fit(house)
```

```python
#cal median for all column
imp.statistics_
```

```python
#Fit new medians in all NaN vals
X = imp.transform(house)
house_tr = pd.DataFrame(X, columns = house.columns)
house_tr.describe()
```

# Sci-kit learn Design
### 3 types of obj
> 1. Estimators - estimates parameters based on datasets. Eg imputer - has fit() - fits data set and calculates internal parameters and transform().
> 2. Transformers - takes input and gives output based on learnings from fit().
> 3. Predictors - Eg. Lin regr model. fit() and predict() are 2 common func. score() evaluates prediction.

# Feature Scaling 
### 2 types:


> 1. Min-Max Scaling (Normalistion): (value - min) / (max - min). Class MinMaxScaler in Sklearn.
> 2. Standardization: (value - mean) / std. Class StandardScaler in sklearn.

# Creating Pipline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler())
])
```

```python
housing_num_tr = my_pipeline.fit_transform(house_tr)
```

```python
housing_num_tr.shape
```

# Selecting Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
#model = LinearRegression()
model = DecisionTreeRegressor()
model.fit(housing_num_tr, house_labels)
```

```python
some_data = house.iloc[:5]
some_labels = house_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

# Evaluating the prediction
### Lin Regr gives RMS error of 4.84 that too large, so we use Decision Tree Regressor.

### Now that gives error 0.0 ie it has over fitted the data.

```python
from sklearn.metrics import mean_squared_error
house_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(house_labels, house_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

# Use cross validation as a better evaluation technique
### If data set has 10 vals 1 2 3...10 it skips 1 and trains for rest 9 then skips 2 and trains rest 9 and so on and then error is calculated.
### Dec tree gives lesser (3.8035) RMS than Lin regr (4.1863) so we use former.

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, house_labels, scoring = "neg_mean_squared_error", cv=10)
# cv means its 10 folds
rmse_scores = np.sqrt(-scores)
```
